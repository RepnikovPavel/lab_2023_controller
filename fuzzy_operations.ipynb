```python
import config
import numpy as np
import matplotlib.pyplot as plt
from RulesMaker.func_det import *
import os
```

```python
ZE = Distrib(make_func("triangle_m_f", [-1 / 3, 0, 1 / 3]), supp_of_func=(-1 / 3, 1 / 3),max_x = 0)
```

```python
NOT_ZE = FNOT(ZE, [-1,1],20)
```

```python
x_=  np.linspace(-1,1,200)
y_ZE = [ZE(el) for el in x_]
y_NOT_ZE =  [NOT_ZE(el) for el in x_]
fig,ax = plt.subplots()
fig.set_size_inches(4,4)
ax.plot(x_,y_ZE,label=r'$p(x)$',linewidth=3)
ax.plot(x_,y_NOT_ZE,label=r'$\neg p(x)$',linewidth=3)
ax.set_xlim(-1,1)
ax.set_ylim(0,ZE(ZE.max_x))
ax.set_title('NOT',fontsize = 30)
leg = ax.legend(prop = { "size": 14 }, loc ="upper left")
fig.savefig(os.path.join(config.task_dir, './FNOT.png'))
# for line in leg.get_lines():
    # line.set_linewidth(1.0)
```
